**iii) Use the plot() function to produce side-by-side boxplots of Outstate versus Private.**

```python
college['Outstate'].plot(kind='kde');
```

```python
plt.boxplot(college['Outstate']);
```

```python
def side_by_side_boxplot(groupby, columns):
    '''
    groupby: column name
    columns: list of column
    '''
    df_melt = college.melt(id_vars = groupby,
                  value_vars = columns,
                  var_name = 'columns')
    
    b = sns.boxplot(data = df_melt,
                hue = groupby, # different colors for different 'Elite'
                x = 'columns',
                y = 'value',
                order = columns)

    plt.title('Boxplot grouped by ' + groupby) # You can change the title here
    plt.show();
```

```python
side_by_side_boxplot('Private', ['Outstate'])
```

**v. Use the hist() function to produce some histograms with differing numbers of bins for a few of the quantitative variables.** 

You may find the following command useful: 

> fig, ax = plt.subplots(nrows=6, ncols=3, figsize=(15, 10)) 

It will divide the print window into 6 rows and 3 columns, so that four plots can be made simultaneously. Modifying the arguments to this function will divide the screen in other ways.

```python
college.describe().T
```

```python
# Select Only Numerical Columns
num_columns = college.select_dtypes(include=np.number).columns
```

```python
fig, ax = plt.subplots(nrows=6, ncols=3, figsize=(15, 10))

i = 0
for row in ax:
    for col in row:
        try:
            col_name = num_columns[i]
            col.hist(college[col_name])
            col.set_title(col_name)
            i = i + 1
        except:
            break

plt.tight_layout()
plt.show();
```
